After doing some cleaning in the data we realize that some rows in the splitted data are repeated due to spelling error, so we look for another method to merge them. This method is saving the data set into two different files, the first is the data with bfn chemical name 1 with respect to their consumed drug and the second file is bnf chemical name two with respect to their consumed drug.

Now, adding the data set into memory, 1st splitted data is df1 and the second splitted data is df2

```python
df1 = pd.read_excel(r'pre1.xlsx')
df2 = pd.read_excel(r'pre2.xlsx')
```

```python
print(df1.shape)
print(df2.shape)
```

Concating the two data set, the new dataset is df.



```python
df =  pd.concat([df1, df2]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.to_excel(r"NIS Drug pres.xlsx")
```

```python
df.info()
```

After merging, there are some empty cells in the dataset, as you can see in the cell above, so we need to drop the empty cells.

```python
df.dropna(inplace = True)
```

Adding the repeated rows in the dataset by bnf chemical name with the corresponding consumed drug(kg)

```python
df.info()
```

```python
df.to_excel(r"NIS Drugs.xlsx")
```

```python
import pandas as pd


# Fill any missing values in the "BNF_chemical_name" column with an empty string
df["BNF_chemical_name"].fillna("", inplace=True)

# Create a dictionary to store the sum of each chemical name
chemical_sum = {}

for index, row in df.iterrows():
    # Get the chemical name from the current row and remove leading/trailing whitespaces
    chemical_name = str(row["BNF_chemical_name"]).strip().upper()

    # Get the cost for the current row
    cost = row["consumed_drug(kg)"]

    # If the chemical name is already in the dictionary, add the cost to its sum
    if chemical_name in chemical_sum:
        chemical_sum[chemical_name] += cost
    else:
        # If the chemical name is not in the dictionary, initialize its sum
        chemical_sum[chemical_name] = cost

# Convert the dictionary to a DataFrame
summed_df = pd.DataFrame(list(chemical_sum.items()), columns=["BNF_chemical_name", "Total_Weight"])

# Save the summed data as a new CSV file
summed_df.head()
```
